```python
import pandas as pd
 
 
# reading the database
data = pd.read_csv("/Users/iramayyub/Downloads/tips.csv")
 
# printing the top 10 rows
display(data.head(10))
```

```python

import pandas as pd
import matplotlib.pyplot as plt
 
 
# Scatter plot with day against tip
plt.scatter(data['day'], data['tip'], c=data['size'],
            s=data['total_bill'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.colorbar()
 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
 
# Scatter plot with day against tip
plt.plot(data['tip'])
plt.plot(data['size'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 

 
# Bar chart with day against tip
plt.bar(data['day'], data['tip'])
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
# Adding the legends
plt.show()
```
