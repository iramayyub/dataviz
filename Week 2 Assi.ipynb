### Q1) Create a freehand "drawing" in matplotlib.  Be it a triangle, square, rectangle, whatever you can think of.

```python
#creating a parallelogram using matplotlib
import matplotlib.pyplot as p
x1,y1 = [-1,3],[5,5]
x2,y2 = [3,5],[5,3]
x3,y3 = [1,5],[3,3]
x4,y4 = [1,-1],[3,5]
p.plot(x1, y1, x2, y2, x3, y3, x4, y4,marker = '+', color="red")
p.show()
```
